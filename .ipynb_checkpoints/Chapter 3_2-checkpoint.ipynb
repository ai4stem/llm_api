## Regression and Classification

<b>회귀(Regression)</b>와 <b>분류(Classification)</b>의 차이점은?<p>
회귀는 어떤 파라미터 사이의 관계식을 도출하는 것이 목적이지만, 분류는 단지 서로 다른 종류의 데이터들을 구분짓거나 묶는 것이 목적이다.

## 일차함수 관계식 찾기

x 변수, y 변수 데이터 만들기 (리스트 객체)

```python
x = [-3,  31,  -11,  4,  0,  22, -2, -5, -25, -14]
y = [ -2,   32,   -10,   5,  1,   23,  -1,  -4, -24,  -13]
print(x)
print(y)
```

그래프 그리기 (matplotlib)

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

#### 판다스 데이터프레임 만들기

```python
import pandas as pd
df = pd.DataFrame({'X':x, 'Y':y})
df.shape
```

#### 머신러닝 - 사이킷런 *Linear Regression* 모형

<b>선형 회귀(linear regression)</b>는 하나 또는 여러 개의 독립 변수(설명 변수)가 종속 변수(반응 변수)에 미치는 영향을 직선(또는 다차원 공간의 초평면) 형태의 함수로 모델링하여, 관측값과 예측값 간의 오차 제곱합을 최소화하는 방법입니다. 이 기법은 입력값과 출력값 사이의 선형적 관계를 가정하며, 각 독립 변수에 대한 회귀 계수(기울기)와 절편을 추정함으로써 미래의 값을 예측하거나 변수 간의 영향력을 해석할 수 있게 해 줍니다. 최소제곱법을 이용해 모수를 추정하며, 잔차의 정규성·등분산성·독립성 같은 가정을 검토해야 하지만, 간단하면서도 해석이 용이해 통계 분석·머신러닝·경제학 등 다양한 분야에서 폭넓게 활용됩니다.

```python
train_features = ['X']
target_cols = ['Y']
X_train = df.loc[:, train_features].values
y_train = df.loc[:, target_cols].values

print(X_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

y=ax+b에서 기울기(a)를 coefficient, 절편(b)을 intercept라고 한다.

```python
lr.coef_, lr.intercept_
```

y = x + 1

```python
print ("기울기: ", lr.coef_[0][0])
print ("y절편: ", lr.intercept_[0])
```

```python
import numpy as np

X_new = np.array(101).reshape(1, 1)
lr.predict(X_new)
```

```python
X_test = np.arange(11, 16, 1).reshape(-1, 1)
X_test
```

```python
y_pred = lr.predict(X_test)
y_pred
```

## 붓꽃 데이터를 활용한 머신러닝 익히기

```python
# 라이브러리 환경
import pandas as pd
import numpy as np

# skleran 데이터셋에서 iris 데이터셋 로딩
from sklearn import datasets
iris = datasets.load_iris()

# iris 데이터셋은 딕셔너리 형태이므로, key 값을 확인
iris.keys()
```

```python
# 설명이 보고 싶다면 아래의 코드를 실행한다.
print(iris['DESCR'])
```

```python
# target 속성의 데이터셋 크기
print("데이터셋 크기:", iris['target'].shape)

# target 속성의 데이터셋 내용
print("데이터셋 내용: \n", iris['target'])
```

```python
# data 속성의 데이터셋 크기
print("데이터셋 크기:", iris['data'].shape)
# data 속성의 데이터셋 내용 (첫 7개 행을 추출)
print("데이터셋 내용: \n", iris['data'][:7, :])
```

#### Sepal: 꽃밫침, Petal: 꽃잎

```python
# data 속성을 판다스 데이터프레임으로 변환
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("데이터프레임의 형태:", df.shape)
df.head()
```

```python
df['sepal length (cm)']
```

```python
# 열(column) 이름을 간결하게 변경
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
# Target 열 추가
df['Target'] = iris['target']
print('데이터셋의 크기: ', df.shape)
df.head()
```

KDE: 커널 밀도 추정<p>
커널 함수: 좌우 대칭이면서 적분값이 1이 되는 함수

```python
# petal_length 값의 분포 - displot 함수 (kde 밀도 함수 그래프)
sns.displot(x='petal_width', kind='kde', data=df)
plt.show()
```

```python
# 품종별 sepal_length 값의 분포 비교
plt.figure()
for target_value in df['Target'].unique():
    subset = df[df['Target'] == target_value]
    sns.kdeplot(
        x=subset['sepal_length'],
        label=str(target_value)
    )

plt.legend(title='Target')
plt.xlabel('sepal_length')
plt.title('Sepal Length KDE by Target')
plt.show()
```

```python
# 나머지 3개 피처 데이터를 한번에 그래프로 출력
for col in ['sepal_width', 'petal_length', 'petal_width']:
    sns.displot(x=col, hue='Target', kind='kde', data=df)
plt.show()
```
